# Реалізація багатошарового персептрона для вирішення задачі класифікації одягу на датасеті Fashion MNIST

Імпортуємо бібліотеки

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

Завантажуємо датасет та створюємо навчальні та валідаційні вибірки

```python
(x_train_full, y_train_full), (x_test, y_test) = fashion_mnist.load_data()

x_train = x_train_full[:50000]
y_train = y_train_full[:50000]

x_val = x_train_full[50000:]
y_val = y_train_full[50000:]
```

Проводимо центрування та нормування вхідних даних

```python
x_train = x_train.astype('float32') / 255.0
x_val = x_val.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
```

Перетворюємо зображення в одновимірні вектори

```python
x_train = x_train.reshape(-1, 28*28)
x_val = x_val.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)
```

Перетворюємо правильні відповіді в one-hot encoding

```python
y_train = to_categorical(y_train)
y_val = to_categorical(y_val)
y_test = to_categorical(y_test)
```

Функція для створення моделі по схемі 784-128-128-10

```python
def create_model():
    model = Sequential()
    model.add(Dense(128, activation='relu', input_shape=(784,)))
    model.add(Dense(128, activation='elu'))
    model.add(Dense(10, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

Ініціалізуємо експерименти навчання за різних параметрів

```python
experiments = [
    {'batch_size': 32, 'epochs': 10},
    {'batch_size': 64, 'epochs': 10},
    {'batch_size': 128, 'epochs': 10},
    {'batch_size': 32, 'epochs': 15},
    {'batch_size': 64, 'epochs': 15},
    {'batch_size': 128, 'epochs': 15},
    {'batch_size': 32, 'epochs': 20},
    {'batch_size': 64, 'epochs': 20},
    {'batch_size': 128, 'epochs': 20},
]
```

Проводимо експерименти навчання за різних параметрів

```python
histories = []
results = []
for exp in experiments:
    name = f"epochs={exp['epochs']}, batch_size={exp['batch_size']}"
    print(f"Починаємо {name}")
    model = create_model()
    history = model.fit(
        x_train, y_train,
        validation_data=(x_val, y_val),
        epochs=exp['epochs'],
        batch_size=exp['batch_size'],
        verbose=0
    )
    histories.append({'name': name, 'history': history})
    final_val_accuracy = history.history['val_accuracy'][-1]
    results.append({
        'epochs': exp['epochs'],
        'batch_size': exp['batch_size'],
        'val_accuracy': final_val_accuracy
    })
```

Будуємо графіки порівняння отриманої точності

```python
plt.figure(figsize=(12, 8))
for h in histories:
    plt.plot(h['history'].history['val_accuracy'], label=h['name'])

plt.title('Порівняння валідаційної точності')
plt.xlabel('Епохи')
plt.ylabel('Валідаційна точність')
plt.legend()
plt.grid(True)
plt.show()
```

Виводимо таблицю з результатами

```python
results_df = pd.DataFrame(results)
print("\nТаблиця результатів:")
print(results_df)
```

Найкраща точність 89.39% була досягнута при навчанні з 20 епохами та батчем 64.

# Візуальний аналіз даних популярності статті

```python
import pandas as pd
import matplotlib.pyplot as plt
import calendar

articles = pd.read_csv("https://github.com/PavelKilko/applied-application-of-neural-networks/raw/master/lab-3/howpop_train.csv")
```

Виведемо транспонований датасет

```python
articles.transpose()
```

Видалимо стовпці views_lognorm, favs_lognorm, comments_lognorm

```python
articles = articles.drop(["views_lognorm", "favs_lognorm", "comments_lognorm"], axis=1)

articles.transpose()
```

Виведемо статичні дані про датасет

```python
articles.describe().transpose()
```

 Перетворимо published в тип datetime

```python
articles["published"] = pd.to_datetime(articles["published"])

articles.transpose()
```

Додамо стовпці year, month, dayofweek, hour

```python
articles["year"] = [d.year for d in articles["published"]]
articles['month'] = [d.month for d in articles["published"]]
articles['dayofweek'] = [d.isoweekday() for d in articles["published"]]
articles['hour'] = [d.hour for d in articles["published"]]

articles.transpose()
```

```python
articles.info()
```

```python
articles.describe().transpose()
```

Побудуємо ретроспективний графік щомісячної кількості статей 

```python
articles_per_month = articles.groupby(['year', 'month']).size().reset_index(name='article_count')

articles_pivot = articles_per_month.pivot(index='month', columns='year', values='article_count')

articles_pivot.plot(kind='bar', figsize=(12, 6), width=0.8)

plt.title('Кількість публікацій за кожен місяць року')
plt.xlabel('Місяць')
plt.ylabel('Кількість публікацій')
plt.xticks(range(12), ['Січень', 'Лютий', 'Березень', 'Квітень', 'Травень', 'Червень', 'Липень', 'Серпень', 'Вересень', 'Жовтень', 'Листопад', 'Грудень'], rotation=0)
plt.legend(title='Рік', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

Відповідно до графіку найбільша кількість публікацій була в березні 2015 року.

Побудуємо графік залежності середньої кількості публікацій від дня тижня на сайтах Хабр (habrahabr.ru) та Гіктаймс (geektimes.ru) за березень 2015 року.

```python
filtered_articles = articles[
    (articles['domain'].isin(['habrahabr.ru', 'geektimes.ru'])) &
    (articles['published'].dt.year == 2015) &
    (articles['published'].dt.month == 3)
]

articles_per_day = filtered_articles.groupby(['dayofweek', 'domain']).size().reset_index(name='article_count')

day_names_ukr = {
    1: 'Понеділок', 2: 'Вівторок', 3: 'Середа', 4: 'Четвер',
    5: 'Пʼятниця', 6: 'Субота', 7: 'Неділя'
}

articles_per_day['dayofweek'] = articles_per_day['dayofweek'].map(day_names_ukr)

days_in_march = pd.date_range(start='2015-03-01', end='2015-03-31')
dayofweek_counts = days_in_march.dayofweek.value_counts().sort_index() + 1

day_occurrences = {1: 5, 2: 5, 3: 4, 4: 4, 5: 4, 6: 4, 7: 5}

articles_per_day['occurrences'] = articles_per_day['dayofweek'].map(lambda x: day_occurrences[list(day_names_ukr.keys())[list(day_names_ukr.values()).index(x)]])

articles_per_day['average_articles'] = articles_per_day['article_count'] / articles_per_day['occurrences']

articles_pivot = articles_per_day.pivot(index='dayofweek', columns='domain', values='average_articles')

articles_pivot = articles_pivot.reindex(['Понеділок', 'Вівторок', 'Середа', 'Четвер', 'Пʼятниця', 'Субота', 'Неділя'])

articles_pivot.plot(kind='bar', figsize=(10, 6), width=0.8)

plt.title('Залежність середньої кількості публікацій від дня тижня на habrahabr.ru та geektimes.ru у березні 2015')
plt.xlabel('День тижня')
plt.ylabel('Середня кількість публікацій')
plt.xticks(rotation=0)
plt.legend(title='Домен', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

З графіку можна стверджувати, що:

1) Пікова середня кількість публікацій припадає на вівторок (60+ на Хабрі, 30+ на Гіктаймс)
2) Середня кількість публікацій на Хабрі зазвичай в двічі більша ніж на Гіктаймс
3) Середня кількість публікацій на Хабрі та Гіктаймс приблизно однакова в суботу.

Побудуємо графіки залежності середньої кількості переглядів та середньої кількості коментарів від години публікації на сайтах Хабр (habrahabr.ru) та Гіктаймс (geektimes.ru)

```python
filtered_articles = articles[
    (articles['domain'].isin(['habrahabr.ru', 'geektimes.ru']))
]

avg_metrics_per_hour = filtered_articles.groupby(['hour', 'domain']).agg(
    avg_views=('views', 'mean'),
    avg_comments=('comments', 'mean')
).reset_index()

views_pivot = avg_metrics_per_hour.pivot(index='hour', columns='domain', values='avg_views')
comments_pivot = avg_metrics_per_hour.pivot(index='hour', columns='domain', values='avg_comments')

plt.figure(figsize=(10, 6))
views_pivot.plot(kind='line', ax=plt.gca(), marker='o')
plt.title('Середня кількість переглядів на годину публікації')
plt.xlabel('Година публікації')
plt.ylabel('Середня кількість переглядів')
plt.xticks(range(0, 24))
plt.legend(title='Домен', loc='upper left')
plt.grid(True)
plt.tight_layout()

plt.figure(figsize=(10, 6))
comments_pivot.plot(kind='line', ax=plt.gca(), marker='o')
plt.title('Середня кількість коментарів на годину публікації')
plt.xlabel('Година публікації')
plt.ylabel('Середня кільксть коментарів')
plt.xticks(range(0, 24))
plt.legend(title='Домен', loc='upper left')
plt.grid(True)
plt.tight_layout()

plt.show()
```

З графіків можна стверджувати, що:

1) Пікова середня кількість переглядів на Хабрі припадає на 6ту годину ранку (22000+), а на Гіктаймс на 4ту годину ранку (18000+)

2) Пікова середня кількість коментарів на Хабрі припадає на 4ту годину ранку (~50), а на Гіктаймс на 3тю годину ранку (65+)

3) Середня кількість переглядів на Хабрі зазвичай більша за Гіктаймс, але середня кількість коментарів на Гіктаймс завжди більша за Хабр.

4) Середня кількість переглядів на Хабрі та Гіктаймс однакова о 8й годині ранку.

5) На обох сайтах переглядають та коментують більше ранкові статті ніж вечірні. Денні статті переглядають та коментують менше ніж ранкові або вечірні.

Знайдемо статті та час їх публікації, які мають найбільшу кількість переглядів на Хабрі та Гіктаймс.

```python
filtered_articles = articles[
    (articles['domain'].isin(['habrahabr.ru', 'geektimes.ru']))
]

max_views_per_domain = filtered_articles.loc[
    filtered_articles.groupby('domain')['views'].idxmax(), ['domain', 'title', 'views', 'hour']
]

print("Стаття з найбільшою кількістю переглядів для кожного домену:")
for index, row in max_views_per_domain.iterrows():
    print(f"Домен: {row['domain']}")
    print(f"Назва: {row['title']}")
    print(f"Кількість переглядів: {row['views']}")
    print(f"Година публікації: {row['hour']}\n")
```

Знайдемо статті та час їх публікації, які мають найбільшу кількість коментарів на Хабрі та Гіктаймс.

```python
filtered_articles = articles[
    (articles['domain'].isin(['habrahabr.ru', 'geektimes.ru']))
]

max_comments_per_domain = filtered_articles.loc[
    filtered_articles.groupby('domain')['comments'].idxmax(), ['domain', 'title', 'comments', 'hour']
]

print("\nСтаття з найбільшою кількістю коментарів для кожного домену:")
for index, row in max_comments_per_domain.iterrows():
    print(f"Домен: {row['domain']}")
    print(f"Назва: {row['title']}")
    print(f"Кількість коментарів: {row['comments']}")
    print(f"Година публікації: {row['hour']}\n")
```

Побудуємо графіки кількості голосів для топ-20 авторів (популярними будемо вважати авторів які мають найбільшу загальну кількість переглядів за всі статті) на Хабрі та Гіктаймс.

```python
filtered_authors = articles.dropna(subset=['author', 'views', 'votes_plus', 'votes_minus', 'domain'])

author_stats = filtered_authors.groupby(['author', 'domain']).agg(
    total_views=('views', 'sum'),
    total_plus_votes=('votes_plus', 'sum'),
    total_minus_votes=('votes_minus', 'sum')
).reset_index()

top_20_habrahabr = author_stats[author_stats['domain'] == 'habrahabr.ru'].sort_values(by='total_views', ascending=False).head(20)
top_20_geektimes = author_stats[author_stats['domain'] == 'geektimes.ru'].sort_values(by='total_views', ascending=False).head(20)

plt.figure(figsize=(12, 8))
plt.barh(top_20_habrahabr['author'], top_20_habrahabr['total_plus_votes'], color='green', label='Позитивні голоси')
plt.barh(top_20_habrahabr['author'], -top_20_habrahabr['total_minus_votes'], color='red', label='Негативні голоси')
plt.xlabel('Голоси')
plt.ylabel('Автори')
plt.title('Топ-20 авторів на habrahabr.ru та їх позитивні та негативні голоси')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
plt.barh(top_20_geektimes['author'], top_20_geektimes['total_plus_votes'], color='green', label='Позитивні голоси')
plt.barh(top_20_geektimes['author'], -top_20_geektimes['total_minus_votes'], color='red', label='Негативні голоси')
plt.xlabel('Голоси')
plt.ylabel('Автори')
plt.title('Топ-20 авторів на geektimes.ru та їх позитивні та негативні голоси')
plt.legend()
plt.tight_layout()
plt.show()
```

З графіків можна стверджувати, що найбільшу кількість позитивних (120000+ на Хабрі та 100000+ на Гіктаймс) та найбільшу кількість негативних голосів (20000+ на Хабрі та 20000+ на Гіктаймс) має автор @alizar

Побудуємо графіки залежності середньої кількості публікацій на Хабрі та Гіктаймс від години дня по суботах та понеділках

```python
filtered_articles = articles[(articles['dayofweek'].isin([1, 6])) & (articles['domain'].isin(['habrahabr.ru', 'geektimes.ru']))]

articles_per_hour = filtered_articles.groupby(['domain', 'dayofweek', 'hour']).size().reset_index(name='article_count')

articles_pivot_saturday = articles_per_hour[articles_per_hour['dayofweek'] == 6].pivot(index='hour', columns='domain', values='article_count')
articles_pivot_monday = articles_per_hour[articles_per_hour['dayofweek'] == 1].pivot(index='hour', columns='domain', values='article_count')

plt.figure(figsize=(12, 6))
articles_pivot_saturday.plot(kind='bar', ax=plt.gca())
plt.title('Залежність середньої кількісті публікацій від години дня по суботах')
plt.xlabel('Година дня')
plt.ylabel('Середня кількість публікацій')
plt.xticks(range(0, 24), rotation=0)
plt.legend(title='Домен')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
articles_pivot_monday.plot(kind='bar', ax=plt.gca())
plt.title('Залежність середньої кількісті публікацій від години дня по понеділках')
plt.xlabel('Година дня')
plt.ylabel('Середня кількість публікацій')
plt.xticks(range(0, 24), rotation=0)
plt.legend(title='Домен')
plt.tight_layout()
plt.show()
```

З графіків можна стверджувати, що:

1) Пікова кількість публікацій на Хабрі по суботах припадає на 15ту годину вечора (350+)

2) Пікова кількість публікацій на Гіктаймс по суботах припадає на 12ту годину вечора (150+)

3) Пікова кількість публікацій на Хабрі по понеділках припадає на 12ту годину ранку (1400+)

4) Пікова кількість публікацій на Гіктаймс по понеділках припадає на 14ту годину вечора (400+)

5) По понеділках в основному пишуть вдень, а по суботах в основному пишуть ближче до вечора та рано вранці.

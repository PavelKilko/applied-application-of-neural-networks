# Тренування моделі розпізнавання шрифтів на аугментованому контрастному датасеті

```python
import pandas as pd
import numpy as np
import scipy.ndimage
import tensorflow as tf
from tensorflow.keras import regularizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
def rotate_image(image, angle=10):
    image_reshaped = image.reshape(20, 20) 
    rotated_image = scipy.ndimage.rotate(image_reshaped, angle, reshape=False)
    clipped_image = np.clip(rotated_image, 0, 255)
    return clipped_image.astype(np.uint8).flatten()

def shift_image(image, shift_x=2, shift_y=2):
    image_reshaped = image.reshape(20, 20).astype(np.float32)
    shifted_image = scipy.ndimage.shift(image_reshaped, [shift_y, shift_x])
    clipped_image = np.clip(shifted_image, 0, 255)
    return clipped_image.astype(np.uint8).flatten()

def scale_image(image, zoom_factor=1.1):
    image_reshaped = image.reshape(20, 20).astype(np.float32)
    zoomed_image = scipy.ndimage.zoom(image_reshaped, zoom_factor)[:20, :20]
    clipped_image = np.clip(zoomed_image, 0, 255)
    return clipped_image.astype(np.uint8).flatten()

def add_gaussian_noise(image, mean=0, std=5):
    image = image.astype(np.float32) 
    noise = np.random.normal(mean, std, image.shape)
    noisy_image = image + noise
    clipped_image = np.clip(noisy_image, 0, 255)  
    return clipped_image.astype(np.uint8).flatten()

def augment_images(df):
    augmented_rows = []
    for _, row in df.iterrows():
        font = row['font']
        image = row.drop('font').values.astype(np.float32)

        rotated = rotate_image(image, angle=np.random.uniform(-15, 15))
        shifted = shift_image(image, shift_x=np.random.randint(-5, 4), shift_y=np.random.randint(-5, 5))
        scaled = scale_image(image, zoom_factor=np.random.uniform(0.9, 1.1))повнозв'язну
        noisy = add_gaussian_noise(image)
    
        augmented_rows.append([font] + rotated.tolist()) 
        augmented_rows.append([font] + shifted.tolist())
        augmented_rows.append([font] + scaled.tolist())
        augmented_rows.append([font] + noisy.tolist())

    augmented_df = pd.DataFrame(augmented_rows, columns=df.columns)
    augmented_df = augmented_df.replace([np.inf, -np.inf], 0).fillna(0)

    pixel_columns = augmented_df.columns.drop('font')
    augmented_df[pixel_columns] = augmented_df[pixel_columns].astype(np.uint8)
    
    return augmented_df
```

```python
train_url = "https://github.com/PavelKilko/applied-application-of-neural-networks/raw/refs/heads/master/lab-6-7/final/train_dataset.csv"
test_url = "https://github.com/PavelKilko/applied-application-of-neural-networks/raw/refs/heads/master/lab-6-7/final/test_dataset.csv"
```

```python
train = pd.read_csv(train_url)
test = pd.read_csv(test_url)
```

```python
def preprocess_data(train_path, test_path, threshold_value=128):
    train_dataset = pd.read_csv(train_path)
    test_dataset = pd.read_csv(test_path)

    X_train = train_dataset.drop('font', axis=1).values
    y_train = train_dataset['font'].values

    X_test = test_dataset.drop('font', axis=1).values
    y_test = test_dataset['font'].values

    X_train = np.where(X_train >= threshold_value, 255, 0)
    X_test = np.where(X_test >= threshold_value, 255, 0)
    
    X_train = X_train / 255.0
    X_test = X_test / 255.0

    mean_train = np.mean(X_train, axis=0)
    X_train_centered = X_train - mean_train
    X_test_centered = X_test - mean_train

    unique_labels = np.unique(y_train)
    label_to_index = {label: idx for idx, label in enumerate(unique_labels)}
    y_train_encoded = np.array([label_to_index[label] for label in y_train])
    y_test_encoded = np.array([label_to_index[label] for label in y_test])

    num_classes = len(unique_labels)

    y_train_one_hot = to_categorical(y_train_encoded, num_classes=num_classes)
    y_test_one_hot = to_categorical(y_test_encoded, num_classes=num_classes)

    return X_train_centered, X_test_centered, y_train_one_hot, y_test_one_hot, num_classes
```

```python
train_url = "https://github.com/PavelKilko/applied-application-of-neural-networks/raw/refs/heads/master/lab-6-7/final/full_train_dataset.csv"
test_url = "https://github.com/PavelKilko/applied-application-of-neural-networks/raw/refs/heads/master/lab-6-7/final/full_test_dataset.csv"
```

```python
X_train, X_test, y_train, y_test, num_classes = preprocess_data(train_url, test_url)
```

```python
def build_and_train_model(neuron_count, X_train, y_train, X_test, y_test, num_classes):
    model = Sequential()
    
    model.add(Dense(neuron_count, input_shape=(X_train.shape[1],), activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    history = model.fit(
        X_train, y_train,
        epochs=100,
        validation_data=(X_test, y_test),
        batch_size=256,
        callbacks=[early_stopping],
        verbose=0
    )

    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)

    return model, history, test_loss, test_acc
```

```python
neuron_counts = [100, 250, 500, 1000, 2000, 4000, 8000, 16000, 32000]

results = []

plt.figure(figsize=(20, 12))

for neuron_count in neuron_counts:
    print(f"Training model with {neuron_count} neurons in input layer...")

    model, history, test_loss, test_acc = build_and_train_model(neuron_count, X_train, y_train, X_test, y_test, num_classes)

    results.append({"Neuron Count": neuron_count, "Test Accuracy": test_acc, "Test Loss": test_loss})

    plt.plot(history.history['accuracy'], label=f'Train Accuracy ({neuron_count} neurons)')
    plt.plot(history.history['val_accuracy'], label=f'Val Accuracy ({neuron_count} neurons)')

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.show()

results_df = pd.DataFrame(results)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
ax[0].plot(results_df['Neuron Count'], results_df['Test Accuracy'], marker='o')
ax[0].set_xlabel('Number of Neurons in Input Layer')
ax[0].set_ylabel('Test Accuracy')
ax[0].set_title('Test Accuracy vs. Number of Neurons in Input Layer')
ax[0].grid(True)

ax[1].plot(results_df['Neuron Count'], results_df['Test Loss'], marker='o', color='red')
ax[1].set_xlabel('Number of Neurons in Hidden Layer')
ax[1].set_ylabel('Test Loss')
ax[1].set_title('Test Loss vs. Number of Neurons in Input Layer')
ax[1].grid(True)

plt.tight_layout()
plt.show()

print(results_df)
```

```python
# Demonstration of deleting of NaN values
demo = diamonds.copy()
demo
```

```python
# Adding some NaN values in demonstration dataset
import numpy as np

np.random.seed(42)

nan_prob = 0.15 # 15% chance that some value in dataset will be NaN

for i in range(demo.shape[0]):
    for j in range(demo.shape[1]):
        if np.random.rand() <= nan_prob:
            demo.iloc[i, j] = np.NaN
        
```

```python
demo
```

```python
# remove all rows that have at least one NaN value
demo_cleared = demo.dropna()
```

```python
demo_cleared
```

```python
# replace NaN with some "template" value
demo_filled = demo.fillna({
    'carat': demo['carat'].mean(), # or 0.0
    'cut': 'Unknown',
    'color': 'Unknown',
    'clarity': 'Unknown',
    'depth': demo['depth'].mean(), # or 0.0
    'table': demo['table'].mean(), # or 0.0
    'price': demo['price'].mean(), # or 0.0
    'x': demo['x'].mean(), # or 0.0
    'y': demo['y'].mean(), # or 0.0
    'z': demo['z'].mean() # or 0.0
})
```

```python
demo_filled
```

```python
# adding attribute combinations
diamonds["area"] = diamonds["x"] * diamonds["y"]
diamonds["volume"] = diamonds["x"] * diamonds["y"] * diamonds["z"]
```

```python
# describe values of datasets
diamonds.describe()
```

```python
# check values distributions of dataset

%matplotlib inline

import matplotlib.pyplot as plt

diamonds.hist(bins=25, figsize=(20, 15))

plt.show()
```

```python
# split into test and train samples

from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(diamonds, test_size=0.2, random_state=42)
```

```python
test_set.head()
```

```python
diamonds["price"].hist()
```

```python
# categorize diamonds by their price

diamonds["price_cat"] = pd.cut(diamonds["price"], bins=[0, 3750, 7500, 11250, 15000, np.inf], labels=[1, 2, 3, 4, 5])

diamonds["price_cat"].hist()
```

```python
# make stratified sample

from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(diamonds, diamonds["price_cat"]):
    strat_train_set = diamonds.loc[train_index]
    strat_test_set = diamonds.loc[test_index]
```

```python
strat_test_set["price_cat"].value_counts() / len(strat_test_set)
```

```python
diamonds["price_cat"].value_counts() / len(diamonds)
```

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("price_cat", axis=1, inplace=True)
```

```python
diamonds = strat_train_set.copy()
```

```python
# visualize size data

%matplotlib notebook

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the points
scatter = ax.scatter(diamonds['x'], diamonds['y'], diamonds['z'], 
                     s=diamonds["carat"] * 100, label="carat",
                     c=diamonds["price"], cmap="jet", 
                     alpha=0.2)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot of Diamonds Data')

ax.set_xlim(0, 11)
ax.set_ylim(0, 60)
ax.set_zlim(0, 32)

cbar = fig.colorbar(scatter, ax=ax, label='Price')

plt.legend()

# Show plot
plt.show()
```

```python
# looking for correlations

# preparation for ordinal parameters
cut_mapping = {'Fair': 1, 'Good': 2, 'Very Good': 3, 'Premium': 4, 'Ideal': 5}
color_mapping = {'D': 1, 'E': 2, 'F': 3, 'G': 4, 'H': 5, 'I': 6, 'J': 7}
clarity_mapping = {'IF': 1, 'VVS1': 2, 'VVS2': 3, 'VS1': 4, 'VS2': 5, 'SI1': 6, 'SI2': 7, 'I1': 8}

diamonds['cut'] = diamonds['cut'].map(cut_mapping)
diamonds['color'] = diamonds['color'].map(color_mapping)
diamonds['clarity'] = diamonds['clarity'].map(clarity_mapping)

corr_matrix = diamonds.corr()

corr_matrix["price"].sort_values(ascending=False)
```

Paramater "price" has high correlation with parameters "carat", "area", "volume", "x", "y", "z" and low correlation with parameters "color", "clarity", "table".

```python
# check scatter matrix

%matplotlib inline

from pandas.plotting import scatter_matrix

attributes = ["price", "carat", "area", "volume", "x", "y", "z", "table"]

scatter_matrix(diamonds[attributes], figsize=(15, 15))

plt.show()
```
